```python
import numpy as np
import scanpy as sc
import glob
```

```python
files = glob.glob('/dfs/user/qhwang/bio_combo/logs_Steinhart_combo/dummy_summary*/sampled_genes_*.npy')
```

```python
all_loaded = []
```

```python
for f in files:
    loaded = np.load(f)
    all_loaded.extend(loaded)
    #all_loaded.extend([np.sort(x).tolist() for x in loaded])
```

```python
all_loaded_dict = {}

for x in all_loaded:
    try:
        all_loaded_dict[str(x)] += 1
    except:
        all_loaded_dict[str(x)] = 1
```

```python
import pandas as pd
```

```python
results_df = pd.DataFrame.from_dict(all_loaded_dict, orient='index')
```

```python
results_df.sort_values(0, ascending=False)[:50]
```
